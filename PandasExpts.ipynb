# Expt 1: Analyzing Electrical Load Data

```python
# @title
import pandas as pd

# Load the dataset into a Pandas DataFrame
data = {'voltage': [220, 225, 230, 235, 240],
        'current': [10, 12, 11, 13, 10],
        'power': [2200, 2700, 2530, 3055, 2400]}
df = pd.DataFrame(data)

# Calculate descriptive statistics
stats = df.describe()
print(stats)
# Visualize the data
df.plot(title='Electrical Load Data', xlabel='Time', ylabel='Value')
```
